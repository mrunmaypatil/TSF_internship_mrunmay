```python
import seaborn as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("globalterrorismdb_0718dist.csv")
```

```python
pd.set_option('max_columns',df.shape[1]+1)
df
```

```python
df.shape
```

```python
df.describe()
```

```python
data=df[['iyear','imonth','iday']].copy()
data.columns=['year','month','day']
data=data.applymap(str)
df['time']=pd.to_datetime(data[['year','month','day']],format='%Y%m%d',errors='coerce')
df['imonth']=df['imonth'].replace(0,np.nan)
df['iday']=df['iday'].replace(0,np.nan)
df.shape
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.shape[0]
```

```python
null_col=[]
for i in df.columns.values:
    if (df[i].isnull().sum())*100/df.shape[0]>=90:
        null_col.append(i)
print(len(null_col))
print(null_col)
```

```python
all_values=[]
for i in df.columns.values:
    if (df[i].isnull().sum())*100/df.shape[0]==0:
        all_values.append(i)
print(len(all_values))
print(all_values)
```

```python
x=df.isnull().sum()
x=pd.DataFrame(x)
x['features']=df.columns.values
plt.figure(figsize=(20,20))
sb.barplot(x=x[0],y=x['features'])
plt.xlabel('no. of NULL values')
```

```python
plt.figure(figsize=(20,20))
sb.heatmap(df.corr())
```

```python
new_data=df[['iyear','imonth','iday','time','country_txt','nkill','region','city','attacktype1_txt','nwoundte']]
pd.set_option('max_columns',new_data.shape[1]+1)
new_data
```

```python
naming={'iyear':'year','imonth':'month','country_txt':'country','nkill':'kill','nwoundte':'wound','attacktype1_txt':'attacktype'}
new_data.rename(columns=naming,inplace=True)
new_data
```

```python
fig,ax3=plt.subplots(figsize=(40,30))
new_data.hist(ax=ax3)
```

```python
new_data[['kill','wound','year']].groupby(['year']).sum().plot(kind='bar',figsize=(20,10))
```

```python
new_data[['kill','wound','country']].groupby(['country']).sum().sort_values(['kill','wound'],axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10))
```

```python
new_data[['kill','country']].groupby(['country']).sum().sort_values(['kill'],axis=0,ascending=False).head(1)
```

```python
new_data[['kill','wound','attacktype']].groupby(['attacktype']).sum().plot(kind='bar',figsize=(20,10))
```

```python
pd.crosstab(new_data.year,new_data.region).plot(kind='area',figsize=(16,8))
```

```python
plt.figure(figsize=(18,5))
sb.countplot('year',data=new_data)
plt.xticks(rotation=90)
```

```python
attacks_percent=new_data['year'].value_counts().to_dict()
y=np.round(((attacks_percent[2017]-attacks_percent[1970])/attacks_percent[2017])*100)
print(f"The total increase in attacks from 1970 to 2017 has been {y}%")
```

```python
new_data['attacktype'].value_counts().plot(kind='bar',figsize=(15,6),color="Yellowgreen")
```
