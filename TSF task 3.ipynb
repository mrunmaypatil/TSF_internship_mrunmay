```python
features=['Segment','Region','Category','Ship Mode']
```

```python
plt.figure(figsize=(15, 30))  # Setting the size of the figure to be ploted
for i, col in enumerate(features):
    plt.subplot(len(features),2 , i+1)
    sns.countplot(col,data=data)
    plt.title(f"Count per {col}")
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data.Sales,shade=True,color="blue")
sns.kdeplot(data.Profit,shade=True,color="red")
plt.xlim(-100,1000)
```

```python
pie_counts=list(data['Category'].value_counts())
my_labels=['Office Supplies','Furniture','Technology']
```

```python
plt.figure(figsize=(8,8))
myexplode = [0.2, 0.1, 0.1]
plt.pie(pie_counts,labels=my_labels,explode=myexplode,shadow=True,autopct='%1.1f%%')
plt.legend()
```
